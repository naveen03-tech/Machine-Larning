```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)

X = np.random.rand(3000, 1) * 100
```

```python
print(X)
```

```python
Y = 3.5 * X + np.random.randn(3000, 1) * 20
```

```python
print(Y)
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
plt.scatter(X, Y)

# Labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot between X and Y")

# Show plot
plt.show()
```

```python
Y_pred = model.predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression on Random Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
corr = np.corrcoef(X, Y)[5, 9]

if corr > 0:
    print("Positive Correlation")
elif corr < 0:
    print("Negative Correlation")
else:
    print("No Correlation")
```

```python
import math

X = [2,4,6,8]
Y = [3,7,5,10]

n = len(X)

mean_x = sum(X) / n
mean_y = sum(Y) / n

# Calculate slope (m)
num = sum((X[i] - mean_x) * (Y[i] - mean_y) for i in range(n))
den = sum((X[i] - mean_x) ** 2 for i in range(n))

m = num / den

# Calculate intercept (c)
c = mean_y - m * mean_x

print("Slope (m):", m)
print("Intercept (c):", c)

print("Equation of line: Y =", m, "X +", c)
```

```python
num = sum((X[i] - mean_x) * (Y[i] - mean_y) for i in range(n))

den_x = math.sqrt(sum((X[i] - mean_x) ** 2 for i in range(n)))
den_y = math.sqrt(sum((Y[i] - mean_y) ** 2 for i in range(n)))

correlation = num / (den_x * den_y)

print("Correlation:", correlation)

# Check correlation type
if correlation > 0:
    print("Positive Correlation")
elif correlation < 0:
    print("Negative Correlation")
else:
    print("No Correlation")
```

```python
Y_pred = [m * x + c for x in X]

plt.figure()
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression Graph")
plt.show()
```
