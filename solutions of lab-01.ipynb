```python
import pandas as pd

# Load CSV file
df = pd.read_csv("insurance.csv")   # change filename if needed

# Calculate correlation matrix (numeric columns only)
corr = df.corr(numeric_only=True)

# Choose target column
target = "charges"   # change if your target column name is different

print("Correlation values with", target)
print(corr[target])

# Check correlation type
positive = corr[target][corr[target] > 0].count()
negative = corr[target][corr[target] < 0].count()

print("\nSummary:")
if positive > negative:
    print("Dataset shows more POSITIVE correlation")
elif negative > positive:
    print("Dataset shows more NEGATIVE correlation")
else:
    print("Dataset has balanced or NO clear correlation")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv('insurance.csv')

# Show first rows (optional)
print(df.head())

# Take columns
labels = df['charges']
values = df['age']

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart from CSV Data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('insurance.csv')

print(df.head())
print(df.columns)   # see actual column names

# Count categories
counts = df['sex'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Sex Distribution')
plt.axis('equal')
plt.show()
```
